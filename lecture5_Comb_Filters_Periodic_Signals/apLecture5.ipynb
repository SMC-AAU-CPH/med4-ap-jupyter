## Comb filtering
In the next 20 minutes, you will learn
- what a feedforward comb filter is
- whan a feedback comb filter is
- what the impulse and frequency responses are of these filters.

<center>
<img src="figures/comb.jpg" alt="Comb" width="60%"/>
</center>


In a comb filter, we have a single delay element delaying the input signal or output signal by D samples.
<center>
<img src="figures/ap5CombFilterStructures.png" alt="Comb filter structures" width="80%"/>
</center>

### Feedforward comb filter
We form the output signal as
$$
    y_n = x_n+b x_{n-D}
$$
where
- $x_n$ is the **input signal**
- $y_n$ is the **output signal**
- $D$ is the **delay** (in samples)
- $b<0$ is the **filter coefficient** and (usually) **negative**

Since the output only depends on the input, the filter is a **feedforward filter**.

#### Impulse response of the feedforward comb filter
If we send the impulse
$$
    \delta_n = \begin{cases}
        1 & n=0\\
        0 & \text{otherwise}
    \end{cases}
$$
through the feedforward comb filter.
<center>
<img src="figures/ap5Impulse.png" alt="Impulse" width="80%"/>
</center>

We get the impulse response $h_n = y_n$
$$
    h_n = \begin{cases}
        1 & n=0\\
        b & n=D\\
        0 & \text{otherwise}
    \end{cases}\ .
$$
<center>
<img src="figures/ap5ImpulseResponseFeedforward.png" alt="Feedforward comb filter impulse response" width="80%"/>
</center>

#### Frequency response of the feedforward comb filter
Recall from the last lecture that the frequency response of a feedforward filter is
$$
    H(\omega) = \sum_{m=0}^q b_m \mathrm{e}^{-j\omega m}\ .
$$
This was derived by sending a phasor $\mathrm{e}^{j\omega n}$ through a feedforward filter.

For the feedforward comb filter $y_n = x_n+b x_{n-D}$, we thus have
$$
    H(\omega) = 1+b\mathrm{e}^{-j\omega D}\ .
$$

By using Euler's identity, we obtain
$$
    H(\omega) = 1+b\mathrm{e}^{-j\omega D} = 1+b\cos(\omega D)-jb\sin(\omega D)
$$
from which can derive the **amplitude response** to
$$
    |H(\omega)| = \sqrt{1+b^2+2b\cos(\omega D)}
$$
and the **phase response** to
$$
    \angle H(\omega) = \mathrm{arctan}\left(\frac{-b\sin(\omega D)}{1+b\cos(\omega D)}\right)\ .
$$

```python
# %matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def firCombFreqResp(filterCoefficient, delay, freqGrid):
    return 1+filterCoefficient*np.exp(-1j*delay*freqGrid)
# compute the frequency response
nFreqs = 500
freqGrid = 2*np.pi*np.arange(nFreqs)/nFreqs
filterCoefficient = np.array([-0.8])
delay = 16
freqResp = firCombFreqResp(filterCoefficient, delay, freqGrid)
ampResp = np.abs(freqResp)
```

```python
# plot the results
plt.figure(figsize=(10,6))
plt.plot(freqGrid, ampResp, linewidth=2)
plt.xlim((freqGrid[0],freqGrid[nFreqs-1])), plt.ylim((0,1.1*np.max(ampResp)))
plt.xlabel('Frequency [rad/sample]'), plt.ylabel('Amplitude response [.]');
```

### Feedback comb filter
We form the output signal as
$$
    y_n = x_n+a y_{n-D}
$$
where
- $x_n$ is the **input signal**
- $y_n$ is the **output signal**
- $D$ is the **delay** (in samples)
- $a>0$ is the **filter coefficient** and (usually) **positive**

Since the output depends on previous output samples, the filter is a **feedback filter**.

#### Impulse response of the feedback comb filter
If we send the impulse
$$
    \delta_n = \begin{cases}
        1 & n=0\\
        0 & \text{otherwise}
    \end{cases}
$$
through the feedback comb filter.
<center>
<img src="figures/ap5Impulse.png" alt="Impulse" width="80%"/>
</center>

We get the impulse response $h_n = y_n$
$$
    h_n = \begin{cases}
        a^{n/D} & n=0, D, 2D, 3D, \ldots\\
        0 & \text{otherwise}
    \end{cases}\ .
$$
<center>
<img src="figures/ap5ImpulseResponseFeedback.png" alt="Feedback comb filter impulse response" width="80%"/>
</center>

#### Frequency response of the feedback comb filter
Recall from the last lecture that the frequency response of a feedback filter is
$$
    H(\omega) = \frac{b_0}{1-\sum_{m=1}^p a_m \mathrm{e}^{-j\omega m}}\ .
$$
This was derived by sending a phasor $\mathrm{e}^{j\omega n}$ through a feedback filter.

For the feedback comb filter $y_n = x_n+a y_{n-D}$, we thus have
$$
    H(\omega) = \frac{1}{1-a\mathrm{e}^{-j\omega D}}\ .
$$

By using Euler's identity, we obtain
\begin{align}
    H(\omega) = \frac{1}{1-a\mathrm{e}^{-j\omega D}} = \frac{1-a\cos(\omega D)}{1+a^2-2a\cos(\omega D)}-j\frac{a\sin(\omega D)}{1+a^2-2a\cos(\omega D)}
\end{align}
from which can derive the **amplitude response** to
$$
    |H(\omega)| = \sqrt{\frac{1}{1+a^2-2a\cos(\omega D)}}
$$
and the **phase response** to
$$
    \angle H(\omega) = \mathrm{arctan}\left(\frac{-a\sin(\omega D)}{1-a\cos(\omega D)}\right)\ .
$$

```python
# %matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def iirCombFreqResp(filterCoefficient, delay, freqGrid):
    return 1/(1-filterCoefficient*np.exp(-1j*delay*freqGrid))
# compute the frequency response
nFreqs = 5000
freqGrid = 2*np.pi*np.arange(nFreqs)/nFreqs
filterCoefficient = np.array([0.8])
delay = 16
freqResp = iirCombFreqResp(filterCoefficient, delay, freqGrid)
ampResp = np.abs(freqResp)
```

```python
# plot the results
plt.figure(figsize=(10,6))
plt.plot(freqGrid, ampResp, linewidth=2)
plt.xlim((freqGrid[0],freqGrid[nFreqs-1])), plt.ylim((0,1.1*np.max(ampResp)))
plt.xlabel('Frequency [rad/sample]'), plt.ylabel('Amplitude response [.]');
```

### Summary
1. A comb filter has a comb-like **amplitude spectrum**.
2. The amplitude response of a **feedforward** comb-filter has sharp dips and soft peaks.
3. The amplitude response of a **feedback** comb-filter has sharp peaks and soft dips.

## Periodic signals
In the next 20 minutes, you will learn

A periodic signal is a signal which **repeats itself**.
<center>
<img src="figures/isTheSignalPeriodic0.png" alt="Sinusoid" width="50%"/>
</center>

Mathematically, we can write this as
$$
    x_n = x_{n-N}
$$
for some period $N$.

In lecture 2, we saw that the vibrations on a guitar string were periodic.

A periodic signal repeats its values after a number of samples $N$, i.e., 
$$
    x_n = x_{n-N}\ .
$$

Note that
1. if you know the function values for one period $n\in[n_1-N,n_1)$, say, then you know the function for all times $n\in(-\infty,\infty)$
2. The inverse of the period $N$ is called the **fundamental frequency** (sometimes also **pitch**) of the periodic signal and denoted by $f_0$, i.e., 
$$
    f_0 = 1/N\quad \text{[cycles/sample]}\ .
$$
The fundamental frequency describes how many times the signals $x_n$ repeats itself pr. sample.

#### Fourier series representation of a periodic signal
Every real-valued periodic discrete-time (and bandlimited) signal can be written as a weighted sum of sinusoids plus a DC-value of $a_0/2$. This can be written as
$$
    x_n = a_0/2+\sum_{k=1}^{\infty} A_k\cos(k\omega_0 n + \psi_k)
$$
where
- $a_0/2$ is the DC value
- $A_k\geq0$ is an amplitude
- $\psi_k$ is the initial phase
- $k\omega_0$ is the digital frequency of the $k$ harmonic components.

Since (from Euler's formula)
$$
    \cos(\theta) = \frac{1}{2}\left[\mathrm{e}^{j\theta}+\mathrm{e}^{-j\theta}\right]\ ,
$$
we can also write $x_n$ more compactly as
\begin{align}
    x_n &= \sum_{k=-K}^{K} c_k\mathrm{e}^{j k\omega_0 n}\\
\end{align}
where
\begin{align}
    c_k &= \begin{cases}
        (A_k/2)\mathrm{e}^{j \psi_k} & k=1,\ldots,K\\
        a_0/2 & k=0\\
        c_{-k}^* & k= -K, -K+1, \ldots, -1\\
        \end{cases}\ .
\end{align}

The weights $c_k$ are calculated as
$$
    c_k = \frac{1}{N} \sum_{n=0}^{N-1} x_n \mathrm{e}^{-j k\omega_0 n}\ .
$$
where $\omega_0 = 2\pi/N$.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wave
import IPython.display as ipd

# load a guitar signal
samplingFreq, fullGuitarSignal = wave.read('data/guitar.ff.sulB.B3.wav')
fullGuitarSignal = fullGuitarSignal/2**15 # normalise
ipd.Audio(fullGuitarSignal, rate=samplingFreq) 
```

```python
# setup
tStart = 0.45 # s
tEnd =0.47 #
startIdx = np.int(tStart*samplingFreq)
endIdx = np.int(tEnd*samplingFreq)
time = np.arange(startIdx,endIdx)/samplingFreq
nData = np.size(time)
```

```python
# compute the pitch period (rounded to the nearest integer for a the B3 note (pitch of 246.94 Hz))
pitch = 246.94 # Hz
pitchPeriod = samplingFreq/pitch
integerPitchPeriod = np.int(np.round(pitchPeriod))
print('For the B3 note (i.e., '+ str(pitch) + ' Hz), the pitch period is N=' + str(pitchPeriod) + ' samples.')
```

```python
#extract a segment
dataSegment = fullGuitarSignal[startIdx:endIdx]
segmentShift = integerPitchPeriod
shiftedDataSegment = fullGuitarSignal[startIdx-segmentShift:endIdx-segmentShift]

# plot the data segment
plt.figure(figsize=(10,6))
plt.plot(time,dataSegment,linewidth=2,label="guitar segment")
plt.plot(time,shiftedDataSegment,linewidth=2,label="shifted guitar segment")
plt.plot(time,dataSegment-shiftedDataSegment,linewidth=2,label="difference")
plt.xlim((time[0],time[nData-1]))
plt.ylim((-0.25,0.25))
plt.grid(True)
plt.legend()
plt.xlabel('time [s]')
plt.ylabel('Amplitude [.]');
```

```python
def computeFourierSeriesWeights(data, integerPitchPeriod):
    nHarmonics = np.int(np.floor(integerPitchPeriod/2))
    harmonicIndices = np.arange(-nHarmonics,nHarmonics+1)
    samplingIndices = np.arange(integerPitchPeriod)
    dataSegment = data[samplingIndices]
    weights = np.zeros(2*nHarmonics+1,dtype=np.complex_)
    # compute the DC value
    weights[nHarmonics] = np.mean(dataSegment)
    # compute remaining weights
    for k in np.arange(nHarmonics):
        phasor = np.exp(-1j*2*np.pi*k*samplingIndices/integerPitchPeriod)
        weights[nHarmonics+k] = np.sum(dataSegment*phasor)/integerPitchPeriod
        weights[nHarmonics-k] = np.conj(weights[nHarmonics+k])
    return weights, harmonicIndices
```

```python
# compute the Fourier series weights
fourierWeights, harmonicIndices = computeFourierSeriesWeights(dataSegment, integerPitchPeriod)
absFourierWeights = np.abs(fourierWeights)

# plot the magnitude of the weights
maxFreq2Plot = 15*pitch
plt.figure(figsize=(10,6))
plt.stem(harmonicIndices*pitch, absFourierWeights, use_line_collection="true")
plt.xlim((0,maxFreq2Plot))
plt.xlabel('frequency [Hz]')
plt.ylabel('Magnitude of Fourier series weights');
```

```python
# design feedback comb filter with a delay equal to the pitch period
nFreqs = 5000
freqGrid = 2*np.pi*np.arange(nFreqs)*maxFreq2Plot/(nFreqs*samplingFreq)
filterCoefficient = 0.95
delay = pitchPeriod
freqResp = iirCombFreqResp(filterCoefficient, delay, freqGrid)
ampResp = np.abs(freqResp)*np.max(absFourierWeights)/np.max(np.abs(freqResp))
# plot the comb filter's frequency response
plt.figure(figsize=(10,6))
plt.plot(freqGrid*samplingFreq/(2*np.pi), ampResp, 'r--',label="IIR comb filter")
plt.stem(harmonicIndices*pitch, absFourierWeights, label="Fourier weights", use_line_collection="true")
plt.xlim((0,maxFreq2Plot))
plt.legend()
plt.xlabel('frequency [Hz]')
plt.ylabel('Magnitude of Fourier series weights');
```

```python
def computeSignalFromFourierWeights(weights,integerPitchPeriod, nData):
    nHarmonics = (np.size(weights)-1)/2
    harmonicIndices = np.arange(-nHarmonics,nHarmonics+1)
    signal = np.zeros(nData)
    samplingIndices = np.arange(nData)
    for n in samplingIndices:
        phasor = np.exp(1j*2*np.pi*harmonicIndices*n/integerPitchPeriod)
        # real is only applied for numerical reasons
        signal[n] = np.real(np.sum(weights*phasor))
    return signal, samplingIndices
    
# let us reconstruct the guitar signal now
reconstructedSignal, _ = computeSignalFromFourierWeights(fourierWeights, integerPitchPeriod, nData)
```

```python
# plot the original and reconstructed signals
plt.figure(figsize=(10,6))
plt.plot(time,dataSegment,linewidth=2,label="guitar segment")
plt.plot(time,reconstructedSignal,linewidth=2,label="reconstructed guitar segment")
plt.plot(time,dataSegment-reconstructedSignal,linewidth=2,label="difference")
plt.xlim((time[0],time[nData-1]))
plt.ylim((-0.25,0.25))
plt.grid(True)
plt.legend()
plt.xlabel('time [s]')
plt.ylabel('Amplitude [.]');
```

## Plucked-string synthesis
In the next 20 minutes, you will learn
- how the comb filter can be used to simulate the sound of a plucked string instrument such as a guitar

We have seen that
- a feedback comb filter can be designed to have peaks at the same frequencies as the harmonics of a guitar signal (in general any periodic signal)
- that the amplitudes of the peaks are the **same** in a comb filter, but **different** for the harmonics of a guitar signal
- the amplitudes of the harmonics of a guitar signal are generally **decaying** with increasing frequency

```python
# plot the comb filter's frequency response
plt.figure(figsize=(10,6))
plt.plot(freqGrid*samplingFreq/(2*np.pi), ampResp, 'r--',label="IIR comb filter")
plt.stem(harmonicIndices*pitch, absFourierWeights, label="Fourier weights",use_line_collection="true")
plt.xlim((0,maxFreq2Plot))
plt.legend()
plt.xlabel('frequency [Hz]')
plt.ylabel('Magnitude of Fourier series weights');
```

### Karplus-strong algorithm
Idea: simulate a guitar string by combining a 
- **comb filter**: produces harmonics with the desired frequencies
- **low-pass filter**: produces harmonics with decaying amplitude as a function of frequency
<center>
<img src="figures/ap5KarplusStrongSimple.png" alt="Simple version of Karplus Strong" width="70%"/>
</center>

From the figure, we obtain
\begin{align}
    w_n &= x_n + ay_{n-D}\\
    y_n &= 0.5w_n+0.5w_{n-1}
\end{align}
<center>
<img src="figures/ap5KarplusStrongSimple.png" alt="Simple version of Karplus Strong" width="60%"/>
</center>

It can be shown that the frequency response of the combined comb and lowpass filter is
$$
    H(\omega) = \frac{0.5+0.5\mathrm{e}^{-j\omega}}{1-\alpha(\omega)\mathrm{e}^{-j\omega(D+0.5)}}
$$
where
$$
    \alpha(\omega) = a\cos(\omega/2)\ .
$$

We can also write the frequency response as
$$
    H(\omega) = H_\text{FIR}(\omega)H_\text{IIR}(\omega)
$$
where
\begin{align}
    H_\text{FIR}(\omega) &= 0.5+0.5\mathrm{e}^{-j\omega}\\
    H_\text{IIR}(\omega) &= \frac{1}{1-\alpha(\omega)\mathrm{e}^{-j\omega(D+0.5)}}
\end{align}
which we recognise as the frequency responses of a **feedforward (FIR)** filter and a **feedback (IIR)** filter, respectively (see first block of this lecture).

Note that
- the delay in the feedback comb filter is $D+0.5$ instead of $D$ !
- we have to take this extra delay of $0.5$ into account if we want to get the 'right' frequencies.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def karplusStrongFirFreqResp(freqGrid):
    return 0.5+0.5*np.exp(-1j*freqGrid)

def karplusStrongIirFreqResp(filterCoefficient, delay, freqGrid):
    modFilterCoefficient = filterCoefficient*np.cos(freqGrid/2)
    phasor = np.exp(-1j*freqGrid*(delay+0.5))
    return 1./(1-modFilterCoefficient*phasor)
```

```python
# compute the frequency response of the Karplus-Strong filter
samplingFreq = 44100 # Hz
filterCoefficient = 0.9
pitch = 440 # Hz
pitchPeriod = samplingFreq/pitch
# we substract 0.5 sample to compensate for the added 1/2 in the KS algorithm
modPitchPeriod = pitchPeriod-1/2
integerPitchPeriod = np.int(np.round(modPitchPeriod))
delay = integerPitchPeriod
nFreqs = 20000
freqGrid = 2*np.pi*np.arange(nFreqs)/nFreqs
ksIirFreqResp = karplusStrongIirFreqResp(filterCoefficient, delay, freqGrid)
ksFirFreqResp = karplusStrongFirFreqResp(freqGrid)
```

```python
# plot the result
plt.figure(figsize=(10,6))
plt.plot(freqGrid*samplingFreq/(2*np.pi), np.abs(ksIirFreqResp), linewidth=2, label="Comb filter")
plt.plot(freqGrid*samplingFreq/(2*np.pi), np.abs(ksFirFreqResp)/(1-filterCoefficient), linewidth=2, label="Lowpass filter")
plt.plot(freqGrid*samplingFreq/(2*np.pi), np.abs(ksIirFreqResp*ksFirFreqResp), linewidth=2, label="Combined")
plt.legend()
plt.xlim((0,samplingFreq/2))
plt.ylim((0,np.max(np.abs(ksIirFreqResp))))
plt.xlabel('frequency [Hz]')
plt.ylabel('Amplitude response');
```

### Fractional delay filters
Assume that we wish to reproduce the pitch $f_0=440$ Hz using the Karplus-Strong algorithm.
- At a sampling frequency $f_\text{s}$ of 44100 Hz, this pitch corresponds to a pitch period $N$ of
$$
    N = \frac{f_\text{s}}{f_0} = 100.2273\text{ samples.}
$$
- This pitch period is **not** an integer which we is required in our current implementation.
- If we round it to the nearest integer, i.e., set $N=100$, this corresponds to a pitch of 441 Hz.
- Thus, we need a way to implement **fractional delays**.

A fractional delay can be implemented using an **allpass filter** given by
$$
    y_{n} = b x_n + x_{n-1} - by_{n-1}\ . 
$$

This allpass filter has frequency response
$$
    H(\omega) = \frac{b+\mathrm{e}^{-j\omega}}{1+b\mathrm{e}^{-j\omega}}
$$
with amplitude and phase responses given by
\begin{align}
    |H(\omega)| &= 1\\
    \angle H(\omega) &= -2\text{arctan}\left[\frac{1-b}{1+b}\tan(\omega/2)\right] \approx -\frac{1-b}{1+b}\omega\ .
\end{align}

Note the following:
- The filter is called an **allpass filter** since its amplitude response is 1.
- The approximation is valid when both $\omega$ and $\angle H(\omega)$ are small. This is typically satisfied in plucked string synthesis.
- The phase response controls the delay through the filter. If we want a delay of $d$ samples, we have to choose $b$ so that
$$
    d \approx \frac{1-b}{1+b}\ .
$$
If we solve this for $b$, we obtain
$$
    b \approx \frac{1-d}{1+d}
$$

### Karplus-Strong algorithm with allpass filter
The Karplus-Strong algorithm can now be written as
\begin{align}
    w_n &= x_n+ay_{n-D}\\
    v_n &= 0.5w_n+0.5w_{n-1}\\
    y_n &= bv_n+v_{n-1}-by_{n-1}\ .
\end{align}
This is illustrated in the figure below.
<center>
<img src="figures/ap5KarplusStrongAllpass.png" alt="Simple version of Karplus Strong" width="80%"/>
</center>

To implement the Karplus-Strong algorithm, we have to specify three parameters.
- The delay $D$: Assume we wish to reproduce the pitch $f_0$ using a sampling frequency of $f_\text{s}$. This corresponds to the pitch period
$$
    N = \frac{f_\text{s}}{f_0}
$$
which is typically non-integer. The delay $D$ is set as the largest integer smaller than $N-0.5$, i.e.,
$$
    D = \lfloor N-0.5 \rfloor\ .
$$
- The allpass parameter $b$: This parameter is computed from the desired fractional delay $d$ which is given by
$$
    d = N-D-0.5\ .
$$
From this, we parameter is calculated as
$$
    b = \frac{1-d}{1+d}\ .
$$
- The comb filter parameter $a$: Is typically set to a value close to, but not exceeding 1.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd

def karplusStrongSyntheziser(inputSignal, combDelay, combParameter, apParameter, nData):
    samplingIndices = np.arange(nData)
    outputSignal = np.zeros(nData)
    lpFilterState = np.zeros(1)
    apFilterState = np.zeros(2)
    for n in samplingIndices:
        # compute lp-filter input
        if n < combDelay:
            lpInput = inputSignal[n]
        else:
            lpInput = inputSignal[n]+combParameter*outputSignal[n-combDelay]
        # compute ap-filter input
        apInput = 0.5*lpInput+0.5*lpFilterState
        lpFilterState = lpInput
        # compute filter output
        outputSignal[n] = apParameter*(apInput-apFilterState[1])+apFilterState[0]
        apFilterState[0] = apInput
        apFilterState[1] = outputSignal[n]
    return outputSignal, samplingIndices
```

```python
# setup
samplingFreq = 44100 # Hz
filterCoefficient = 0.99
pitch = 110 # Hz
pitchPeriod = samplingFreq/pitch
simulationTime = 3 #s
nData = np.int(simulationTime*samplingFreq)
#inputSignal = np.r_[1,np.zeros(nData-1)] # an impulse
nNoiseSamples = 200
inputSignal = np.r_[np.random.randn(nNoiseSamples),np.zeros(nData-nNoiseSamples)] # a short noise burst


# compute Karplus-Strong filter parameters
combDelay = np.int(np.floor(pitchPeriod-0.5))
fractionalDelay = pitchPeriod-combDelay-0.5
apParameter = (1-fractionalDelay)/(1+fractionalDelay)

# run the algorithm
outputSignal, samplingIndices = \
    karplusStrongSyntheziser(inputSignal, combDelay, filterCoefficient, apParameter, nData)

# play back the signal
ipd.Audio(outputSignal, rate=samplingFreq) 
```

### Summary
1. We can synthesize a guitar signal using a combination of a
 - feedback comb filter
 - lowpass filter
 - allpass filter
2. An allpass filter can be used to make a fractional (i.e., non-integer delay)

### Example: basic musical notation for a standard piano
<br />

<center>
<img src="figures/piano.jpg" alt="Piano illustration" width="80%"/>
</center>

- A standard piano has **88 keys**.
- If a key is pressed down, a (nearly) **periodic signal** is produced.
- The **pitch** of the signal depends on which key is pressed.
 - If the first key (the left-most key) is pressed, the pitch is 27.50 Hz.
 - If the 88th key (the right-most key) is pressed, the pitch is 4186.009 Hz.

- Piano keys are arranged in groups called **octaves** which is divided into 12 different **semitones**.
- A **semitone** is represented with
 - a **letter**: like A, B, C, D, E, F, G
 - and possibly **a symbol**: either $\sharp$ for sharp of $\flat$ for flat
- The 12 semitones are C, C$\sharp=$D$\flat$, D, D$\sharp=$E$\flat$, E, F, F$\sharp=$G$\flat$, G, G$\sharp=$A$\flat$, A, A$\sharp=$B$\flat$, and B
- To make it more clear, we will here also add the octave number to the semitone (e.g.,C$_2\sharp$ means the semitone C$\sharp$ in the second octave).
<center>
<img src="figures/allPianoKeys.png" alt="Piano illustration 2" width="90%"/>
</center>

- The semitone C$_i\sharp$ has a **pitch twice as a big** as the semitone C$_{i-1}\sharp$. This holds in general for all semitones so you
 - double the pitch if you move exactly 12 keys to the right
 - half the pitch if you move exactly 12 keys to the left
- Notes that are one or several octaves apart belong to the same **pitch class** or **chroma**. For example, C$_1\sharp$ and C$_6\sharp$ belong to the same pitch class.

<center>
<img src="figures/piano.jpg" alt="Piano illustration" width="80%"/>
</center>

- If you produce the pitch $f_n$ by pressing key number $n$, you produce the pitch
$$
    f_{n+k} = 2^{k/12}f_n
$$
by pressing on key $n+k$ instead. Note that (as described in the previous cell)
 - if we go an octave up, i.e., set $k=12$, then the pitch doubles since $f_{n+12} = 2f_n$
 - if we go an octave down, i.e., set $k=-12$, then the pitch halves since $f_{n-12} = (1/2)f_n$.
- To calibrate the piano, people have agreed to make the note A$_4$ the reference with a value of 440 Hz (usually)

- Since the note A$_4$ corresponds to key number 49 on our piano, the pitch produced by key number $n$ can be written as
$$
    f_n = 2^{\frac{n-49}{12}}f_{49}
$$
where $f_{49} = 440$ Hz.
- If a piano does not produce these frequencies, it is **out of tune**.
- Finally, the frequency range between two semitones can be divided into cents:
 - 100 cents corresponds to the range between two semitones
 - 1200 cents correponds to the range between two octaves

### The MIDI scale
- The Musical Instrument Digital Interface (MIDI) tuning standard assigns a number to every possible pitch.
 - Example: MIDI note 69 corresponds to A$_4$ or 440 Hz.
- In general, the MIDI number $F_0$ can be computed from a pitch $f_0$ as
$$
    F_0 = 69 + 12\log_2\left(\frac{f_0}{440\text{ Hz}}\right)
$$
Note that $F_0$ is an integer so only the pitches of the form
$$
    f_0 = 2^{\frac{k}{12}}440\text{ Hz}
$$
are allowed.
